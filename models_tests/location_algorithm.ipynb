```
import math

def haversine(lat1,len1,lat2,len2):
  """ Calcula la distacia entre dos puntos geogradicos """
  rad = math.pi/180 
  dlat=lat2-lat1
  dlen=len2-len1
  R=6372.795477598

  a=(math.sin(rad*dlat/2))**2 + math.cos(rad*lat1)*math.cos(rad*lat2)*(math.sin(rad*dlen/2))**2
  distancia = 2*R*math.asin(math.sqrt(a))
  return distancia
```
